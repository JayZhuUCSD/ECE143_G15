```python
def main():
    import Counts_genres_in_publisher as cgp
    import DATA_PROCESSING
    import plot_functions_top_1000 as plots
    import processing_functions_for_plots as processing 
    import DATA_PROCESSING
    import Text_cloud as tc
    import zipfile
    
    # Get platform dataframe
    df_platform = processing.countPlatform()
    # Save to csv
    df_platform.to_csv('platform_count.csv')   
    # Plot the platform dataframe
    print('Platform Graph')
    plots.plotPlatform1000()
    
    
    # Get specs dataframe
    df_specs = processing.countQuantity(col_name='specs')
    # Save to csv
    df_specs.to_csv('specs_count.csv') 
    # Plot the specs dataframe
    print('Specs Graph')
    plots.plotSpecs1000()
    
    
    # Get genres dataframe
    df_genres = processing.countQuantity(col_name='genres') 
    # Save to csv
    df_genres.to_csv('genres_count.csv') 
    # Plot the genre dataframe
    print('Genre Graph')
    plots.plotGenre1000()
    
    
    # Get developers dataframe
    df_publishers = processing.countPublishers()
    # Save to csv
    df_publishers.to_csv('publishers_count.csv')
    # Plot the publisher dataframe
    print('Publisher Graph')
    plots.plotPublisher1000()
    
    # Form Publisher and Genre csv
    cgp.CartesianProduct_features()
    cgp.Feature2_in_Feature1()
    # Plot Publisher and Genre Relationship
    print('Publisher and Genre Graph')
    plots.plotGenreToPlatform()
    
    # Unzip txt files of reviews
    with zipfile.ZipFile("text_reviews.zip","r") as zip_ref:
        zip_ref.extractall(".")
        
        # Generate/display text clouds
        tc.generateTextCloud_top10()
        tc.generateTextCloud_ordinary()
    
    
if __name__ == "__main__":
    main()
```
